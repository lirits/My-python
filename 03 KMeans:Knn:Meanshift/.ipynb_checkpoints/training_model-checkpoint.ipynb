# 2D Data classification

1. Using Kmeans function to classification 2D Data
2. Predict the Data [80 , 60]
3. repeat the data training in KNN and Meanshift model

## KMeans model 

```python
# load the data 
import numpy as np
import pandas as pd
data = pd.read_csv('data.csv')
data.head()
```

```python
# define X and y
X = data.drop(['labels'],axis = 1)
y = data.loc[:,'labels']
X.head()
```

```python
# show labels value_counts
pd.value_counts(y)
```

```python
 # visualize original data
from matplotlib import pyplot as plt
fig1 = plt.figure()
plt.scatter(X.loc[:,'V1'],X.loc[:,'V2'])
plt.title('original data')
plt.xlabel('V1')
plt.ylabel('V2')
plt.show()
```

```python
#visualize the original data in label 
fig2 = plt.figure()
label0 = plt.scatter(X.loc[:,'V1'][y==0],X.loc[:,'V2'][y==0])
label1 = plt.scatter(X.loc[:,'V1'][y==1],X.loc[:,'V2'][y==1])
label2 = plt.scatter(X.loc[:,'V1'][y==2],X.loc[:,'V2'][y==2])
plt.title('original data in label')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
plt.show()
```

```python
# set the model
from sklearn.cluster import KMeans
KM = KMeans(n_clusters=3,random_state=1)
KM.fit(X)
```

```python
#visualize the new data wiwh KMeans_centers
centers = KM.cluster_centers_
fig3 = plt.figure()
label0 = plt.scatter(X.loc[:,'V1'][y==0],X.loc[:,'V2'][y==0])
label1 = plt.scatter(X.loc[:,'V1'][y==1],X.loc[:,'V2'][y==1])
label2 = plt.scatter(X.loc[:,'V1'][y==2],X.loc[:,'V2'][y==2])
plt.scatter(centers[:,0],centers[:,1])
plt.title('origins data with KMeans_centers')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
plt.show()
```

```python
# test data : V1=80 V2=60
y_predict_test = KM.predict([[80,60]])
y_predict_test
```

```python
# predict based on training data 
y_predict = KM.predict(X)
print(pd.value_counts(y_predict),pd.value_counts(y))
```

```python
#calculation accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
print(accuracy)
```

```python
# visualize the KMeans_data and original_data 
fig5 = plt.figure()
label0 = plt.scatter(X.loc[:,'V1'][y_predict==0],X.loc[:,'V2'][y_predict==0])
label1 = plt.scatter(X.loc[:,'V1'][y_predict==1],X.loc[:,'V2'][y_predict==1])
label2 = plt.scatter(X.loc[:,'V1'][y_predict==2],X.loc[:,'V2'][y_predict==2])
plt.scatter(centers[:,0],centers[:,1])
plt.title('KMeans model data')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
fig6 = plt.figure()
label0 = plt.scatter(X.loc[:,'V1'][y==0],X.loc[:,'V2'][y==0])
label1 = plt.scatter(X.loc[:,'V1'][y==1],X.loc[:,'V2'][y==1])
label2 = plt.scatter(X.loc[:,'V1'][y==2],X.loc[:,'V2'][y==2])
plt.scatter(centers[:,0],centers[:,1])
plt.title('original data')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
plt.show()
```

```python
#correct the results
y_correct = []
for i in y_predict:
    if i == 0:
        y_correct.append(0)
    elif i ==1:
        y_correct.append(2)
    else:
        y_correct.append(1)
y_correct = np.array(y_correct)
print(pd.value_counts(y_correct))
print(pd.value_counts(y))
```

```python
print(accuracy_score(y,y_correct))
```

```python
# visualize the corrcet KMeans_data 
fig5 = plt.figure()
label0 = plt.scatter(X.loc[:,'V1'][y_correct==0],X.loc[:,'V2'][y_correct==0])
label1 = plt.scatter(X.loc[:,'V1'][y_correct==1],X.loc[:,'V2'][y_correct==1])
label2 = plt.scatter(X.loc[:,'V1'][y_correct==2],X.loc[:,'V2'][y_correct==2])
plt.scatter(centers[:,0],centers[:,1])
plt.title('KMeans_correct_model data')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
fig6 = plt.figure()
label0 = plt.scatter(X.loc[:,'V1'][y==0],X.loc[:,'V2'][y==0])
label1 = plt.scatter(X.loc[:,'V1'][y==1],X.loc[:,'V2'][y==1])
label2 = plt.scatter(X.loc[:,'V1'][y==2],X.loc[:,'V2'][y==2])
plt.scatter(centers[:,0],centers[:,1])
plt.title('original data')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
plt.show()
```

# establish KNN and meanshift model

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(X,y)
```

```python
#predict the data:[80,60]
test = KNN.predict([[80,60]])
test
```

```python
#predict the data and calculate the accuracy score
y_predict_KNN = KNN.predict(X)
KNN_accuracy = accuracy_score(y,y_predict_KNN)
KNN_accuracy
```

```python
%%time
#establish Meansshift model
from sklearn.cluster import MeanShift,estimate_bandwidth
#obtain the bandwidth
bw = estimate_bandwidth(X,n_samples=500)
print(bw)
MS = MeanShift(bandwidth=bw)
MS.fit(X)
```

```python
y_predict_MS = MS.predict(X)
print(pd.value_counts(y_predict_MS))
print(pd.value_counts(y))
```

```python
# correct the results
y_correct_MS = []
for i in y_predict_MS:
    if i == 0:
        y_correct_MS.append(2)
    elif i ==1:
        y_correct_MS.append(1)
    else:
        y_correct_MS.append(0)
y_correct_MS = np.array(y_correct_MS)
print(pd.value_counts(y_correct_MS))
print(pd.value_counts(y))
```

```python
# calculate the accuracy score
accurace_MS = accuracy_score(y,y_correct_MS)
accurace_MS
```

```python
fig6 = plt.figure()
label0 = plt.scatter(X.loc[:,'V1'][y_correct_MS==0],X.loc[:,'V2'][y_correct_MS==0])
label1 = plt.scatter(X.loc[:,'V1'][y_correct_MS==1],X.loc[:,'V2'][y_correct_MS==1])
label2 = plt.scatter(X.loc[:,'V1'][y_correct_MS==2],X.loc[:,'V2'][y_correct_MS==2])
plt.title('Meanshift data')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
plt.show()
```
