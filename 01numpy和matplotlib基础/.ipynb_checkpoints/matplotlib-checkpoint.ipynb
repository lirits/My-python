# matplotlib

```python
import numpy as np
x = np.linspace(1,10,100)
```

```python
y = np.sin(x)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.show()
```

```python
sinX = np.sin(x)
cosX = np.cos(x)
```

```python
plt.plot(x,sinX)
plt.plot(x,cosX)
plt.show()
```

### [matplotlib.color](https://matplotlib.org/2.0.2/api/colors_api.html)

```python
plt.plot(x,sinX)
plt.plot(x,cosX,color='r',linestyle='--')
plt.show()
```

[linestyle](https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html)

```python
plt.plot(x,sinX)
plt.plot(x,cosX,color='r',linestyle='--')
plt.xlim(-5,15)
plt.ylim(0,1)
plt.show()
```

```python
plt.plot(x,sinX)
plt.plot(x,cosX,color='r',linestyle='--')
plt.axis([0,10,-1,1])
plt.show()
```

```python
plt.plot(x,sinX)
plt.plot(x,cosX,color='r',linestyle='--')
plt.axis([0,10,-1,1])
plt.xlabel('sinX')
plt.ylabel('cosX')
plt.show()
```

```python
plt.plot(x,sinX,label='sin(x)')
plt.plot(x,cosX,color='r',linestyle='--',label='sin(X)')
plt.axis([0,10,-1,1])
plt.xlabel('sinX')
plt.ylabel('cosX')
plt.legend()
plt.title('sinX VS cos X')
plt.show()
```

## Scatter Plot 

```python
plt.scatter(x,sinX,color='g')
plt.show()
```

```python
plt.scatter(x,sinX)
plt.scatter(x,cosX,color='r')
plt.show()
```

```python
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)
plt.scatter(x,y)
plt.show()
```

```python
import random
random.seed(3)
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)
plt.scatter(x,y,alpha=0.5)
plt.show()
```
